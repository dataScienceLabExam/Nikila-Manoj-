```python
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('/content/sample_data/car_evaluation.csv',names=['buying price','maintainance cost','no.of doors','no.of persons','lugboot','safety','unaccurate'])
```

```python
data
```

```python
X=data.drop('unaccurate',axis=1)

X
```

```python
y=data['unaccurate']
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, y_train, X_test, y_test= train_test_split(X,y,test_size=0.3,random_state=50)
```

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier()
```

```python
clf.fit(X_train,y_train)
```

```python
y_pred=clf.predict(X_test)

```

```python
print("accuracy:",accuracy_score(y_test,y_pred)
)
```

```python
from sklearn.tree import export_graphviz
import pydotplus
from six import StringIO
from IPython.display import Image
```

```python
dot_data=StringIO()
export_graphviz(clf,out_file=dot_data,filled=True,rounded=True,special_characters=True,features_name=data.feature_name,class_name=data.target_name)
graph=pydotplus.graph_from_dot_data(dot_data.get_value())
Image(graph.create_png())
```
